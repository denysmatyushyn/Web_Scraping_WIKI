```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_ISO_639-2_codes'
page = requests.get(url)
```

```python
soup_obj = BeautifulSoup(page.text, 'lxml')
soup_obj
```

```python
table_iso = soup_obj.find('table', id='iso-codes')
table_iso
```

```python
headers_table = []
for x in table_iso.find_all('th'):
    title_col = x.text
    headers_table.append(title_col)
headers_table    
```

```python
df = pd.DataFrame(columns = headers_table)
for y in table_iso.find_all('tr')[1:]:
    row_data = y.find_all('td')
    row = [y.text for y in row_data]
    df = df.append(pd.Series(row, index=df.columns[:len(row)], dtype=pd.StringDtype()), ignore_index=True)
```
